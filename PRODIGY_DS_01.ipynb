```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("drug200.csv")
df
```

```python
df.head(10)
```

```python
# 1. Age Distribution (Histogram)
plt.figure(figsize=(6, 4))  
plt.hist(df['Age'], bins='auto', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.grid(True)
plt.show()
```

```python
# 2. Blood Pressure Levels (Bar Chart)
plt.figure(figsize=(6, 4))  
bp_counts = df['BP'].value_counts()
bp_counts.plot(kind='bar', color=['blue', 'green', 'pink'])  
plt.xlabel('Blood Pressure Levels (BP)')
plt.ylabel('Count')
plt.title('Blood Pressure Levels Distribution')
plt.grid(True)
plt.show()
```

```python
# 3. Cholesterol Levels (Bar Chart)
plt.figure(figsize=(6, 4))  
chol_counts = df['Cholesterol'].value_counts()
chol_counts.plot(kind='bar', color=['red', 'yellow', 'green'])  
plt.xlabel('Cholesterol Levels')
plt.ylabel('Count')
plt.title('Cholesterol Levels Distribution')
plt.grid(True)
plt.show()
```

```python
# 4. Sex Distribution (Bar Chart)
plt.figure(figsize=(6, 4))  
sex_counts = df['Sex'].value_counts()
sex_counts.plot(kind='bar', color=['blue', 'pink'])  
plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Sex Distribution')
plt.grid(True)
plt.show()
```

```python
# 5. Drug Type Distribution (Bar Chart)
plt.figure(figsize=(6, 4)) 
drug_counts = df['Drug'].value_counts()
drug_counts.plot(kind='bar', color=['purple', 'orange']) 
plt.xlabel('Drug Type')
plt.ylabel('Count')
plt.title('Drug Type Distribution')
plt.grid(True)
plt.show()
```

```python
# Numerical Variables (Histogram)
sns.histplot(data=df, x='Age', kde=True, color='blue')  
plt.title('Age Distribution')
plt.grid(True)
plt.show()
```

```python
# Categorical Variables (Bar Chart with custom color palette)
sns.countplot(x='Sex', data=df, palette=['skyblue', 'pink'])  
plt.title('Sex Distribution')
plt.grid(True)
plt.show()
```

```python
sns.countplot(x='BP', data=df, palette=['purple', 'orange', 'green'])
plt.title('Blood Pressure Levels Distribution')
plt.grid(True)
plt.show()
```

```python
sns.countplot(x='Cholesterol', data=df, palette=['red', 'yellow', 'green'])
plt.title('Cholesterol Levels Distribution')
plt.grid(True)
plt.show()
```

```python
sns.countplot(x='Drug', data=df, palette=['purple', 'orange'])
plt.title('Drug Type Distribution')
plt.grid(True)
plt.show()
```

# Histograms with Hue for Relationships

```python
# Age vs. Blood Pressure Levels
sns.histplot(x='Age', hue='BP', data=df, element='step', common_norm=False)  
plt.title('Age Distribution by Blood Pressure Levels')
plt.grid(True)
plt.show()
```

```python
# Age vs. Cholesterol Levels
sns.histplot(x='Age', hue='Cholesterol', data=df, element='step', common_norm=False)  
plt.title('Age Distribution by Cholesterol Levels')
plt.grid(True)
plt.show()
```
